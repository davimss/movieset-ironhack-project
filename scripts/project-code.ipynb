```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Passo 3: Definir as categorias de gênero
genres = [
    'Drama', 'History', 'Adventure', 'Fantasy', 'Biography', 'Romance',
    'Crime', 'War', 'Mystery', 'Horror', 'Western', 'Comedy', 'Family',
    'Action', 'Sci-Fi', 'Thriller', 'Sport', 'Musical', 'Music', 'Film-Noir',
    'Animation', 'Adult', 'Documentary', 'Reality-TV', 'Game-Show', 'News'
]

# Passo 4: Criar colunas binárias para cada gênero
for genre in genres:
    column_name = f'{genre.lower()}_movies'
    df[column_name] = df['genre'].apply(lambda genres: 1 if genre in genres else 0)

# Passo 5: Preparar os dados para clustering
# Selecionar apenas as colunas binárias para o clustering
X = df[[f'{genre.lower()}_movies' for genre in genres]]

# Opcional: Normalizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Passo 6: Aplicar o modelo de clustering (K-means como exemplo)
kmeans = KMeans(n_clusters=5, random_state=42)  # Escolha o número de clusters adequado
df['cluster'] = kmeans.fit_predict(X_scaled)

# Passo 7: Exibir resultados do clustering
print(df[['title', 'cluster']].head())

# Passo 8: Visualizar os clusters (opcional, se o número de características for reduzido)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['cluster'], cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clusters of Movies')
plt.colorbar(label='Cluster')
plt.show()
```

```python
df = df
cols = []
for i in range(len(df.columns)):
    cols.append(df.columns[i].lower().replace('-', '_'))
df.columns = cols
df
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Convert 'year' column from int to str
df['year'] = df['year'].astype(str)
df['duration'] = df['duration'].astype(str)


# Substituir NaNs por strings vazias
df['genre'] = df['genre'].fillna('')
df['director'] = df['director'].fillna('')
df['actors'] = df['actors'].fillna('')
df['title'] = df['title'].fillna('')
df['language'] = df['language'].fillna('')
df['year'] = df['year'].fillna('')
df['duration'] = df['duration'].fillna('')

# Criar a coluna combinada
df['combined_features'] = df['genre'] + " " + df['director'] + " " + df['actors']  + " " + df['title'] + " " + df['language'] + " " + df['year'] + " " + df['duration']	

# Passo 3: Converter texto em vetores numéricos
vectorizer = TfidfVectorizer(stop_words='english')
feature_vectors = vectorizer.fit_transform(df['combined_features'])

# Passo 4: Criar a função de recomendação
def recommend_movies(movie_title, num_recommendations=10):
    movie_index = df[df['title'] == movie_title].index[0]
    similarity_scores = cosine_similarity(feature_vectors[movie_index], feature_vectors)
    similar_movies = list(enumerate(similarity_scores[0]))
    sorted_similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)[1:num_recommendations+1]
    recommended_titles = []
    for i in sorted_similar_movies:
        recommended_titles.append(df.iloc[i[0]]['title'])
    
    return recommended_titles
```
